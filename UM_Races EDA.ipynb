### Let us now extract the records for USA in 50Km & 50 miles within the year 2020.

```python
# step 1: Show 50 miles or 50 Kms
df[df['Event distance/length'].isin(['50km', '50mi'])]
```

```python
# Step 2: let us now move on and do further filtering to extract record of 50km or 50 miles within year 2020
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020)]
```

#### Step 3: To get the overall results we want, it is still required to combine the above results with races organized in USA
##### In our data we do not have any columns which specifically tells us in which country the event was organized. So to filter out the races organized in USA, we will use the "Event name" column to extract country names.

```python
Unique_val = df['Event name'].unique()
print(Unique_val)
```

```python
# now to check what code is used for united states in the "Event name" columns. We used google and searched about the races
# organized in USA. From our results we picked one 'Everglades 50 Mile Ultra Run' and we now use this to check the code for united sts

df[df['Event name'] == 'Everglades 50 Mile Ultra Run (USA)']
```

```python
# to extract 'USA' from the 'Event column' we use str.split() and place it in new column country_code. See below code line
df['country_code'] = df['Event name'].str.split('(').str[-1].str.split(')').str[0]
df
```

```python
# now we will combine the code in cell 8 and cell 11 to get the required result.
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['country_code'] == 'USA')]
```

```python
# let us store these results within a new dataframe
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['country_code'] == 'USA')]
df2
```

```python
# Now let us clean up this df2. First we will remove (USA) from 'Event name column'
df2['Event name'] = df2['Event name'].str.split('(').str.get(0) 
```

```python
df2
```

```python
df2['athlete_age'] = 2020 - df['Athlete year of birth']
df2
```

```python
# removing 'h' from athlete performance
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
df2
```

```python
# dropping columns: 'Athelete Club', 'Athlete Country', 'Athlete Year of birth', 'Athlete age category'
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
df2.head(10)
```

```python
# clean up null values. Let us check how many null values we have
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna() == 1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
# checking duplicates
df2[df2.duplicated() == True]
```

```python
# resetting index
df2.reset_index(drop= True)
```

```python
df2.dtypes
```

```python
# changing dtypes
df2['athlete_age'] = df2['athlete_age'].astype(int)
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
# renaming columns
df2 = df2.rename(columns= {'Year of event': 'year',
                           'Event dates': 'race_day', 
                           'Event name': 'race_name',
                           'Event distance/length': 'race_length',
                           'Event number of finishers': 'race_number_of_finishers',
                           'Athlete performance': 'athlete_performance',
                           'Athlete gender': 'athlete_gender',
                           'Athlete average speed': 'athlete_average_speed',
                           'Athlete ID': 'athlete_id'})
df2.head(10)
```

```python
# reorder columns 
df3 = df2[['race_day', 'race_name','race_length','race_number_of_finishers', 'athlete_id', 'athlete_gender', 'athlete_age','athlete_performance','athlete_average_speed', 'country_code']]
df3.head(10)
```

### Problems
#### P1: Difference in speed for the 50km, 50mi male to female

```python
df3.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

#### P2: What age groups are the best in the 50m Race (20+ races minimum)

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending= False).query('count>19')
```

#### P3: Race Seasons

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3.head(5)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df3.head()
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending= False)
```
